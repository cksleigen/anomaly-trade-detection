```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import koreanize_matplotlib
koreanize_matplotlib.koreanize()  # 이 줄 추가!


import seaborn as sns
from datetime import timedelta
import warnings
warnings.filterwarnings('ignore')

class NormalTradingPatternAnalyzer:
    def __init__(self, data_dict):
        """
        data_dict: {'Reward': df, 'Funding': df, 'Trade': df, 'IP': df, 'Spec': df}
        """
        self.reward_df = data_dict['Reward'].copy()
        self.funding_df = data_dict['Funding'].copy()
        self.trade_df = data_dict['Trade'].copy()
        self.ip_df = data_dict['IP'].copy()
        self.spec_df = data_dict['Spec'].copy()

        # 시간 컬럼 변환
        self._prepare_data()

    def _prepare_data(self):
        """데이터 전처리"""
        # 시간 컬럼 변환
        for df_name in ['reward_df', 'funding_df', 'trade_df']:
            df = getattr(self, df_name)
            if 'ts' in df.columns:
                df['ts'] = pd.to_datetime(df['ts'])
                df['date'] = df['ts'].dt.date
                df['hour'] = df['ts'].dt.hour
                df['day_of_week'] = df['ts'].dt.dayofweek

        if 'day' in self.spec_df.columns:
            self.spec_df['day'] = pd.to_datetime(self.spec_df['day'])

    def analyze_all_patterns(self):
        """모든 정상 패턴 분석 실행"""
        print("=" * 80)
        print("정상 거래 패턴 분석 시작")
        print("=" * 80)

        results = {}

        # 1. 거래 빈도 패턴
        print("\n[1] 거래 빈도 패턴 분석 중...")
        results['trading_frequency'] = self.analyze_trading_frequency()

        # 2. 거래 규모 패턴
        print("\n[2] 거래 규모 패턴 분석 중...")
        results['trading_volume'] = self.analyze_trading_volume()

        # 3. 포지션 관리 패턴
        print("\n[3] 포지션 관리 패턴 분석 중...")
        results['position_management'] = self.analyze_position_management()

        # 4. 수익성 패턴
        print("\n[4] 수익성 패턴 분석 중...")
        results['profitability'] = self.analyze_profitability()

        # 5. 다각화 패턴
        print("\n[5] 다각화 패턴 분석 중...")
        results['diversification'] = self.analyze_diversification()

        # 6. 계정 행동 패턴
        print("\n[6] 계정 행동 패턴 분석 중...")
        results['account_behavior'] = self.analyze_account_behavior()

        # 7. 시간대별 패턴
        print("\n[7] 시간대별 패턴 분석 중...")
        results['temporal_pattern'] = self.analyze_temporal_pattern()

        print("\n" + "=" * 80)
        print("분석 완료!")
        print("=" * 80)

        return results

    def analyze_trading_frequency(self):
        """거래 빈도 패턴 분석"""
        account_stats = []

        for account_id, group in self.trade_df.groupby('account_id'):
            # 기본 통계
            total_trades = len(group)
            trading_days = group['date'].nunique()
            avg_daily_trades = total_trades / trading_days if trading_days > 0 else 0

            # 시간당 거래 횟수
            hourly_trades = group.groupby('hour').size()
            avg_hourly_trades = hourly_trades.mean()
            max_hourly_trades = hourly_trades.max()

            # 거래 간격 (초 단위)
            trade_intervals = group.sort_values('ts')['ts'].diff().dt.total_seconds()
            avg_interval = trade_intervals.mean()
            min_interval = trade_intervals.min()

            account_stats.append({
                'account_id': account_id,
                'total_trades': total_trades,
                'trading_days': trading_days,
                'avg_daily_trades': avg_daily_trades,
                'avg_hourly_trades': avg_hourly_trades,
                'max_hourly_trades': max_hourly_trades,
                'avg_interval_seconds': avg_interval,
                'min_interval_seconds': min_interval
            })

        freq_df = pd.DataFrame(account_stats)

        # 정상 패턴 요약
        summary = {
            'total_accounts': len(freq_df),
            'avg_trades_per_account': freq_df['total_trades'].mean(),
            'median_trades_per_account': freq_df['total_trades'].median(),
            'avg_daily_trades': freq_df['avg_daily_trades'].mean(),
            'avg_interval_seconds': freq_df['avg_interval_seconds'].mean(),
            'percentiles': {
                'trades_25': freq_df['total_trades'].quantile(0.25),
                'trades_50': freq_df['total_trades'].quantile(0.50),
                'trades_75': freq_df['total_trades'].quantile(0.75),
                'trades_90': freq_df['total_trades'].quantile(0.90),
            }
        }

        print(f"총 계정 수: {summary['total_accounts']}")
        print(f"평균 거래 횟수/계정: {summary['avg_trades_per_account']:.2f}")
        print(f"평균 일일 거래 횟수: {summary['avg_daily_trades']:.2f}")
        print(f"평균 거래 간격: {summary['avg_interval_seconds']:.1f}초")

        return {'detail': freq_df, 'summary': summary}

    def analyze_trading_volume(self):
        """거래 규모 패턴 분석"""
        account_stats = []

        for account_id, group in self.trade_df.groupby('account_id'):
            # 거래 금액 통계
            avg_amount = group['amount'].mean()
            median_amount = group['amount'].median()
            std_amount = group['amount'].std()
            max_amount = group['amount'].max()

            # 레버리지 통계
            avg_leverage = group[group['openclose'] == 'OPEN']['leverage'].mean()
            max_leverage = group['leverage'].max()

            # 수량 통계
            avg_qty = group['qty'].mean()

            account_stats.append({
                'account_id': account_id,
                'avg_amount': avg_amount,
                'median_amount': median_amount,
                'std_amount': std_amount,
                'max_amount': max_amount,
                'cv_amount': std_amount / avg_amount if avg_amount > 0 else 0,  # 변동계수
                'avg_leverage': avg_leverage,
                'max_leverage': max_leverage,
                'avg_qty': avg_qty
            })

        volume_df = pd.DataFrame(account_stats)

        summary = {
            'avg_trade_amount': volume_df['avg_amount'].mean(),
            'median_trade_amount': volume_df['median_amount'].median(),
            'avg_leverage': volume_df['avg_leverage'].mean(),
            'percentiles': {
                'amount_25': volume_df['avg_amount'].quantile(0.25),
                'amount_50': volume_df['avg_amount'].quantile(0.50),
                'amount_75': volume_df['avg_amount'].quantile(0.75),
                'amount_90': volume_df['avg_amount'].quantile(0.90),
            }
        }

        print(f"평균 거래 금액: ${summary['avg_trade_amount']:.2f}")
        print(f"중앙값 거래 금액: ${summary['median_trade_amount']:.2f}")
        print(f"평균 레버리지: {summary['avg_leverage']:.2f}x")

        return {'detail': volume_df, 'summary': summary}

    def analyze_position_management(self):
        """포지션 관리 패턴 분석"""
        account_stats = []

        for account_id, group in self.trade_df.groupby('account_id'):
            # 포지션별 분석
            position_groups = group.groupby('position_id')

            position_durations = []
            position_profits = []

            for position_id, pos_group in position_groups:
                pos_sorted = pos_group.sort_values('ts')

                # 포지션 보유 시간
                if len(pos_sorted) >= 2:
                    duration = (pos_sorted.iloc[-1]['ts'] - pos_sorted.iloc[0]['ts']).total_seconds()
                    position_durations.append(duration)

                    # 손익 계산 (OPEN, CLOSE 있는 경우)
                    opens = pos_sorted[pos_sorted['openclose'] == 'OPEN']
                    closes = pos_sorted[pos_sorted['openclose'] == 'CLOSE']

                    if len(opens) > 0 and len(closes) > 0:
                        open_amount = opens['amount'].sum()
                        close_amount = closes['amount'].sum()
                        side = opens.iloc[0]['side']

                        if side == 'LONG':
                            profit = close_amount - open_amount
                        else:  # SHORT
                            profit = open_amount - close_amount

                        position_profits.append(profit)

            # OPEN/CLOSE 비율
            open_count = len(group[group['openclose'] == 'OPEN'])
            close_count = len(group[group['openclose'] == 'CLOSE'])
            open_close_ratio = open_count / close_count if close_count > 0 else np.nan

            # 통계 계산
            avg_duration = np.mean(position_durations) if position_durations else 0
            median_duration = np.median(position_durations) if position_durations else 0

            account_stats.append({
                'account_id': account_id,
                'total_positions': position_groups.ngroups,
                'avg_position_duration_seconds': avg_duration,
                'median_position_duration_seconds': median_duration,
                'open_close_ratio': open_close_ratio,
                'completed_positions': len(position_profits)
            })

        position_df = pd.DataFrame(account_stats)

        summary = {
            'avg_positions_per_account': position_df['total_positions'].mean(),
            'avg_position_duration_hours': position_df['avg_position_duration_seconds'].mean() / 3600,
            'avg_open_close_ratio': position_df['open_close_ratio'].mean()
        }

        print(f"평균 포지션 수/계정: {summary['avg_positions_per_account']:.2f}")
        print(f"평균 포지션 보유 시간: {summary['avg_position_duration_hours']:.2f}시간")
        print(f"평균 OPEN/CLOSE 비율: {summary['avg_open_close_ratio']:.2f}")

        return {'detail': position_df, 'summary': summary}

    def analyze_profitability(self):
        """수익성 패턴 분석"""
        account_stats = []

        for account_id, group in self.trade_df.groupby('account_id'):
            position_results = []

            for position_id, pos_group in group.groupby('position_id'):
                pos_sorted = pos_group.sort_values('ts')

                opens = pos_sorted[pos_sorted['openclose'] == 'OPEN']
                closes = pos_sorted[pos_sorted['openclose'] == 'CLOSE']

                if len(opens) > 0 and len(closes) > 0:
                    open_amount = opens['amount'].sum()
                    close_amount = closes['amount'].sum()
                    side = opens.iloc[0]['side']

                    if side == 'LONG':
                        profit = close_amount - open_amount
                        profit_rate = (profit / open_amount * 100) if open_amount > 0 else 0
                    else:
                        profit = open_amount - close_amount
                        profit_rate = (profit / open_amount * 100) if open_amount > 0 else 0

                    position_results.append({
                        'profit': profit,
                        'profit_rate': profit_rate,
                        'is_win': profit > 0
                    })

            if position_results:
                results_df = pd.DataFrame(position_results)

                win_rate = results_df['is_win'].mean()
                avg_profit = results_df['profit'].mean()
                avg_profit_rate = results_df['profit_rate'].mean()

                wins = results_df[results_df['is_win']]
                losses = results_df[~results_df['is_win']]

                avg_win = wins['profit'].mean() if len(wins) > 0 else 0
                avg_loss = losses['profit'].mean() if len(losses) > 0 else 0
                profit_factor = abs(avg_win / avg_loss) if avg_loss != 0 else np.nan
            else:
                win_rate = avg_profit = avg_profit_rate = 0
                avg_win = avg_loss = profit_factor = 0

            account_stats.append({
                'account_id': account_id,
                'completed_positions': len(position_results),
                'win_rate': win_rate,
                'avg_profit': avg_profit,
                'avg_profit_rate': avg_profit_rate,
                'avg_win': avg_win,
                'avg_loss': avg_loss,
                'profit_factor': profit_factor
            })

        profit_df = pd.DataFrame(account_stats)

        summary = {
            'avg_win_rate': profit_df['win_rate'].mean(),
            'median_win_rate': profit_df['win_rate'].median(),
            'avg_profit_rate': profit_df['avg_profit_rate'].mean(),
            'avg_profit_factor': profit_df['profit_factor'].mean()
        }

        print(f"평균 승률: {summary['avg_win_rate']*100:.2f}%")
        print(f"평균 수익률: {summary['avg_profit_rate']:.2f}%")
        print(f"평균 손익비: {summary['avg_profit_factor']:.2f}")

        return {'detail': profit_df, 'summary': summary}

    def analyze_diversification(self):
        """다각화 패턴 분석"""
        account_stats = []

        for account_id, group in self.trade_df.groupby('account_id'):
            # 거래 심볼 수
            unique_symbols = group['symbol'].nunique()
            total_trades = len(group)

            # 심볼별 거래 비중
            symbol_distribution = group['symbol'].value_counts(normalize=True)
            concentration = (symbol_distribution ** 2).sum()  # Herfindahl Index

            # LONG/SHORT 비율
            side_counts = group['side'].value_counts()
            long_ratio = side_counts.get('LONG', 0) / total_trades if total_trades > 0 else 0

            account_stats.append({
                'account_id': account_id,
                'unique_symbols': unique_symbols,
                'symbol_concentration': concentration,
                'long_ratio': long_ratio,
                'most_traded_symbol': group['symbol'].mode()[0] if len(group) > 0 else None,
                'most_traded_symbol_ratio': symbol_distribution.iloc[0] if len(symbol_distribution) > 0 else 0
            })

        div_df = pd.DataFrame(account_stats)

        summary = {
            'avg_unique_symbols': div_df['unique_symbols'].mean(),
            'avg_concentration': div_df['symbol_concentration'].mean(),
            'avg_long_ratio': div_df['long_ratio'].mean()
        }

        print(f"평균 거래 심볼 수: {summary['avg_unique_symbols']:.2f}")
        print(f"평균 집중도: {summary['avg_concentration']:.3f}")
        print(f"평균 LONG 비율: {summary['avg_long_ratio']*100:.2f}%")

        return {'detail': div_df, 'summary': summary}

    def analyze_account_behavior(self):
        """계정 행동 패턴 분석"""
        account_stats = []

        # IP 패턴
        ip_counts = self.ip_df.groupby('account_id')['ip'].nunique()

        # 보상 패턴
        reward_stats = self.reward_df.groupby('account_id').agg({
            'reward_amount': ['count', 'mean', 'sum']
        }).reset_index()
        reward_stats.columns = ['account_id', 'reward_count', 'avg_reward', 'total_reward']

        # 펀딩 패턴
        funding_stats = self.funding_df.groupby('account_id').agg({
            'funding_fee': ['count', 'mean', 'sum']
        }).reset_index()
        funding_stats.columns = ['account_id', 'funding_count', 'avg_funding_fee', 'total_funding_fee']

        # 통합
        trade_accounts = self.trade_df['account_id'].unique()

        for account_id in trade_accounts:
            unique_ips = ip_counts.get(account_id, 0)

            reward_row = reward_stats[reward_stats['account_id'] == account_id]
            reward_cnt = reward_row['reward_count'].values[0] if len(reward_row) > 0 else 0
            avg_reward = reward_row['avg_reward'].values[0] if len(reward_row) > 0 else 0

            funding_row = funding_stats[funding_stats['account_id'] == account_id]
            funding_cnt = funding_row['funding_count'].values[0] if len(funding_row) > 0 else 0
            avg_funding = funding_row['avg_funding_fee'].values[0] if len(funding_row) > 0 else 0

            account_stats.append({
                'account_id': account_id,
                'unique_ips': unique_ips,
                'reward_count': reward_cnt,
                'avg_reward': avg_reward,
                'funding_count': funding_cnt,
                'avg_funding_fee': avg_funding
            })

        behavior_df = pd.DataFrame(account_stats)

        summary = {
            'avg_unique_ips': behavior_df['unique_ips'].mean(),
            'single_ip_ratio': (behavior_df['unique_ips'] == 1).mean(),
            'avg_rewards_per_account': behavior_df['reward_count'].mean(),
            'avg_funding_events': behavior_df['funding_count'].mean()
        }

        print(f"평균 고유 IP 수: {summary['avg_unique_ips']:.2f}")
        print(f"단일 IP 사용 비율: {summary['single_ip_ratio']*100:.2f}%")
        print(f"평균 보상 횟수: {summary['avg_rewards_per_account']:.2f}")

        return {'detail': behavior_df, 'summary': summary}

    def analyze_temporal_pattern(self):
        """시간대별 패턴 분석"""
        # 시간대별 거래 분포
        hourly_dist = self.trade_df.groupby('hour').size()

        # 요일별 거래 분포
        daily_dist = self.trade_df.groupby('day_of_week').size()

        # 가장 활발한 거래 시간대
        peak_hours = hourly_dist.nlargest(3).index.tolist()

        summary = {
            'peak_trading_hours': peak_hours,
            'hourly_distribution': hourly_dist.to_dict(),
            'daily_distribution': daily_dist.to_dict(),
            'most_active_day': int(daily_dist.idxmax()),
            'least_active_day': int(daily_dist.idxmin())
        }

        print(f"피크 거래 시간대: {peak_hours}")
        print(f"가장 활발한 요일: {summary['most_active_day']} (0=월요일)")

        return {'summary': summary}

    def visualize_patterns(self, results):
        """패턴 시각화"""
        fig, axes = plt.subplots(3, 3, figsize=(18, 15))
        fig.suptitle('정상 거래 패턴 분석', fontsize=16, y=1.00)

        # 1. 거래 빈도 분포
        ax = axes[0, 0]
        freq_df = results['trading_frequency']['detail']
        ax.hist(freq_df['total_trades'], bins=30, edgecolor='black', alpha=0.7)
        ax.set_xlabel('총 거래 횟수')
        ax.set_ylabel('계정 수')
        ax.set_title('계정별 거래 횟수 분포')
        ax.axvline(freq_df['total_trades'].median(), color='r', linestyle='--', label='중앙값')
        ax.legend()

        # 2. 거래 금액 분포
        ax = axes[0, 1]
        volume_df = results['trading_volume']['detail']
        ax.hist(np.log10(volume_df['avg_amount'] + 1), bins=30, edgecolor='black', alpha=0.7)
        ax.set_xlabel('평균 거래 금액 (log10)')
        ax.set_ylabel('계정 수')
        ax.set_title('계정별 평균 거래 금액 분포')

        # 3. 레버리지 분포
        ax = axes[0, 2]
        ax.hist(volume_df['avg_leverage'].dropna(), bins=30, edgecolor='black', alpha=0.7)
        ax.set_xlabel('평균 레버리지')
        ax.set_ylabel('계정 수')
        ax.set_title('계정별 평균 레버리지 분포')

        # 4. 포지션 보유 시간
        ax = axes[1, 0]
        position_df = results['position_management']['detail']
        ax.hist(position_df['avg_position_duration_seconds'] / 3600, bins=30, edgecolor='black', alpha=0.7)
        ax.set_xlabel('평균 포지션 보유 시간 (시간)')
        ax.set_ylabel('계정 수')
        ax.set_title('포지션 보유 시간 분포')

        # 5. 승률 분포
        ax = axes[1, 1]
        profit_df = results['profitability']['detail']
        ax.hist(profit_df['win_rate'] * 100, bins=30, edgecolor='black', alpha=0.7)
        ax.set_xlabel('승률 (%)')
        ax.set_ylabel('계정 수')
        ax.set_title('계정별 승률 분포')
        ax.axvline(profit_df['win_rate'].median() * 100, color='r', linestyle='--', label='중앙값')
        ax.legend()

        # 6. 심볼 다각화
        ax = axes[1, 2]
        div_df = results['diversification']['detail']
        ax.hist(div_df['unique_symbols'], bins=range(1, int(div_df['unique_symbols'].max())+2),
                edgecolor='black', alpha=0.7)
        ax.set_xlabel('거래 심볼 수')
        ax.set_ylabel('계정 수')
        ax.set_title('계정별 거래 심볼 수 분포')

        # 7. IP 개수 분포
        ax = axes[2, 0]
        behavior_df = results['account_behavior']['detail']
        ax.hist(behavior_df['unique_ips'], bins=range(1, int(behavior_df['unique_ips'].max())+2),
                edgecolor='black', alpha=0.7)
        ax.set_xlabel('고유 IP 수')
        ax.set_ylabel('계정 수')
        ax.set_title('계정별 IP 개수 분포')

        # 8. 시간대별 거래 분포
        ax = axes[2, 1]
        hourly_dist = pd.Series(results['temporal_pattern']['summary']['hourly_distribution'])
        ax.bar(hourly_dist.index, hourly_dist.values, edgecolor='black', alpha=0.7)
        ax.set_xlabel('시간 (hour)')
        ax.set_ylabel('거래 횟수')
        ax.set_title('시간대별 거래 분포')
        ax.set_xticks(range(0, 24, 2))

        # 9. 요일별 거래 분포
        ax = axes[2, 2]
        daily_dist = pd.Series(results['temporal_pattern']['summary']['daily_distribution'])
        days = ['월', '화', '수', '목', '금', '토', '일']
        ax.bar(range(len(daily_dist)), daily_dist.values, edgecolor='black', alpha=0.7)
        ax.set_xlabel('요일')
        ax.set_ylabel('거래 횟수')
        ax.set_title('요일별 거래 분포')
        ax.set_xticks(range(7))
        ax.set_xticklabels(days)

        plt.tight_layout()
        return fig


# 실행 예시
analyzer = NormalTradingPatternAnalyzer(data)
results = analyzer.analyze_all_patterns()

# 시각화
fig = analyzer.visualize_patterns(results)
plt.show()

# 결과를 DataFrame으로 저장하고 싶다면:
# results['trading_frequency']['detail'].to_csv('normal_frequency_pattern.csv', index=False)
# results['trading_volume']['detail'].to_csv('normal_volume_pattern.csv', index=False)
# ...
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
from datetime import timedelta
from typing import Dict, List, Tuple, Optional
import warnings
warnings.filterwarnings('ignore')

class WashTradingDetector:
    """
    Wash Trading (가장거래) 탐지 시스템

    Wash Trading의 주요 특징:
    1. 짧은 시간 내 반복적인 OPEN-CLOSE
    2. 손익이 거의 0에 가까움 (수수료만 손실)
    3. 동일/유사한 금액의 반복 거래
    4. 특정 심볼에 집중
    5. 비정상적으로 높은 거래 빈도
    """

    def __init__(self, data_dict: Dict[str, pd.DataFrame],
                 thresholds: Optional[Dict] = None):
        """
        Args:
            data_dict: 데이터 딕셔너리
            thresholds: 탐지 임계값 설정
        """
        self.trade_df = data_dict['Trade'].copy()
        self.funding_df = data_dict['Funding'].copy()
        self.reward_df = data_dict['Reward'].copy()
        self.ip_df = data_dict['IP'].copy()
        self.spec_df = data_dict['Spec'].copy()

        # 기본 임계값 설정
        self.thresholds = thresholds or {
            'min_trades': 20,                    # 최소 거래 횟수
            'max_position_duration_minutes': 10, # 최대 포지션 보유 시간 (분)
            'max_profit_rate': 0.1,              # 최대 수익률 (%)
            'min_trade_interval_seconds': 5,     # 최소 거래 간격 (초)
            'amount_similarity_threshold': 0.95, # 금액 유사도 (0-1)
            'symbol_concentration': 0.7,         # 심볼 집중도 (0-1)
            'round_trip_ratio': 0.8,             # 왕복거래 비율 (0-1)
            'min_suspicious_score': 60           # 의심 점수 임계값 (0-100)
        }

        self._preprocess_data()

    def _preprocess_data(self):
        """데이터 전처리"""
        # 시간 변환
        self.trade_df['ts'] = pd.to_datetime(self.trade_df['ts'])
        self.trade_df = self.trade_df.sort_values(['account_id', 'ts'])

        # 거래 간격 계산
        self.trade_df['time_diff'] = self.trade_df.groupby('account_id')['ts'].diff()
        self.trade_df['time_diff_seconds'] = self.trade_df['time_diff'].dt.total_seconds()

    def detect_all_patterns(self) -> pd.DataFrame:
        """모든 Wash Trading 패턴 탐지"""
        print("=" * 80)
        print("🔍 Wash Trading 탐지 시작")
        print("=" * 80)

        suspicious_accounts = []

        for account_id in self.trade_df['account_id'].unique():
            account_trades = self.trade_df[self.trade_df['account_id'] == account_id]

            # 최소 거래 횟수 필터
            if len(account_trades) < self.thresholds['min_trades']:
                continue

            # 각 패턴 탐지
            scores = {}
            evidences = {}

            # 1. 빠른 왕복거래 패턴
            score1, evidence1 = self._detect_rapid_round_trips(account_trades)
            scores['rapid_round_trips'] = score1
            evidences['rapid_round_trips'] = evidence1

            # 2. 제로 수익 패턴
            score2, evidence2 = self._detect_zero_profit_pattern(account_trades)
            scores['zero_profit'] = score2
            evidences['zero_profit'] = evidence2

            # 3. 동일 금액 반복 패턴
            score3, evidence3 = self._detect_identical_amount_pattern(account_trades)
            scores['identical_amount'] = score3
            evidences['identical_amount'] = evidence3

            # 4. 심볼 집중 패턴
            score4, evidence4 = self._detect_symbol_concentration(account_trades)
            scores['symbol_concentration'] = score4
            evidences['symbol_concentration'] = evidence4

            # 5. 고빈도 단기 거래 패턴
            score5, evidence5 = self._detect_high_frequency_pattern(account_trades)
            scores['high_frequency'] = score5
            evidences['high_frequency'] = evidence5

            # 6. 가격 충격 없는 대량 거래
            score6, evidence6 = self._detect_no_price_impact(account_trades)
            scores['no_price_impact'] = score6
            evidences['no_price_impact'] = evidence6

            # 종합 의심 점수 계산 (가중 평균)
            weights = {
                'rapid_round_trips': 0.25,
                'zero_profit': 0.20,
                'identical_amount': 0.20,
                'symbol_concentration': 0.10,
                'high_frequency': 0.15,
                'no_price_impact': 0.10
            }

            total_score = sum(scores[k] * weights[k] for k in scores.keys())

            # 임계값 이상인 경우만 저장
            if total_score >= self.thresholds['min_suspicious_score']:
                suspicious_accounts.append({
                    'account_id': account_id,
                    'suspicious_score': total_score,
                    'total_trades': len(account_trades),
                    'unique_symbols': account_trades['symbol'].nunique(),
                    'trading_period_days': (account_trades['ts'].max() - account_trades['ts'].min()).days,
                    **scores,
                    'evidences': evidences
                })

        if not suspicious_accounts:
            print("\n✅ Wash Trading 의심 계정이 발견되지 않았습니다.")
            return pd.DataFrame()

        result_df = pd.DataFrame(suspicious_accounts)
        result_df = result_df.sort_values('suspicious_score', ascending=False)

        print(f"\n⚠️  {len(result_df)}개의 의심 계정 발견!")
        print("=" * 80)

        return result_df

    def _detect_rapid_round_trips(self, trades: pd.DataFrame) -> Tuple[float, Dict]:
        """빠른 왕복거래 패턴 탐지"""
        position_stats = []

        for position_id, pos_trades in trades.groupby('position_id'):
            pos_sorted = pos_trades.sort_values('ts')

            opens = pos_sorted[pos_sorted['openclose'] == 'OPEN']
            closes = pos_sorted[pos_sorted['openclose'] == 'CLOSE']

            if len(opens) > 0 and len(closes) > 0:
                # 포지션 보유 시간 (분)
                duration_minutes = (closes['ts'].min() - opens['ts'].max()).total_seconds() / 60

                position_stats.append({
                    'duration_minutes': duration_minutes,
                    'is_rapid': duration_minutes < self.thresholds['max_position_duration_minutes']
                })

        if not position_stats:
            return 0, {'count': 0, 'ratio': 0}

        stats_df = pd.DataFrame(position_stats)
        rapid_ratio = stats_df['is_rapid'].mean()
        avg_duration = stats_df['duration_minutes'].mean()

        # 점수 계산: 빠른 거래 비율이 높을수록 높은 점수
        score = min(rapid_ratio * 100, 100)

        evidence = {
            'rapid_position_count': int(stats_df['is_rapid'].sum()),
            'total_position_count': len(stats_df),
            'rapid_ratio': round(rapid_ratio, 3),
            'avg_duration_minutes': round(avg_duration, 2)
        }

        return score, evidence

    def _detect_zero_profit_pattern(self, trades: pd.DataFrame) -> Tuple[float, Dict]:
        """제로 수익 패턴 탐지 (손익이 거의 없는 거래)"""
        position_profits = []

        for position_id, pos_trades in trades.groupby('position_id'):
            pos_sorted = pos_trades.sort_values('ts')

            opens = pos_sorted[pos_sorted['openclose'] == 'OPEN']
            closes = pos_sorted[pos_sorted['openclose'] == 'CLOSE']

            if len(opens) > 0 and len(closes) > 0:
                open_avg_price = (opens['price'] * opens['qty']).sum() / opens['qty'].sum()
                close_avg_price = (closes['price'] * closes['qty']).sum() / closes['qty'].sum()

                side = opens.iloc[0]['side']

                if side == 'LONG':
                    profit_rate = ((close_avg_price - open_avg_price) / open_avg_price) * 100
                else:  # SHORT
                    profit_rate = ((open_avg_price - close_avg_price) / open_avg_price) * 100

                position_profits.append({
                    'profit_rate': abs(profit_rate),
                    'is_near_zero': abs(profit_rate) < self.thresholds['max_profit_rate']
                })

        if not position_profits:
            return 0, {'count': 0, 'ratio': 0}

        profits_df = pd.DataFrame(position_profits)
        zero_profit_ratio = profits_df['is_near_zero'].mean()
        avg_profit_rate = profits_df['profit_rate'].mean()

        # 점수: 제로 수익 거래가 많을수록 높은 점수
        score = min(zero_profit_ratio * 100, 100)

        evidence = {
            'near_zero_profit_count': int(profits_df['is_near_zero'].sum()),
            'total_closed_positions': len(profits_df),
            'near_zero_ratio': round(zero_profit_ratio, 3),
            'avg_abs_profit_rate': round(avg_profit_rate, 3)
        }

        return score, evidence

    def _detect_identical_amount_pattern(self, trades: pd.DataFrame) -> Tuple[float, Dict]:
        """동일/유사 금액 반복 패턴 탐지"""
        amounts = trades['amount'].values

        if len(amounts) < 2:
            return 0, {'similar_pairs': 0}

        # 금액을 정렬하여 유사한 금액 찾기
        sorted_amounts = np.sort(amounts)

        # 연속된 금액 간의 유사도 계산
        similarities = []
        for i in range(len(sorted_amounts) - 1):
            if sorted_amounts[i+1] > 0:
                similarity = sorted_amounts[i] / sorted_amounts[i+1]
                if similarity >= self.thresholds['amount_similarity_threshold']:
                    similarities.append(similarity)

        similar_ratio = len(similarities) / (len(amounts) - 1) if len(amounts) > 1 else 0

        # 가장 빈번한 금액 찾기
        amount_counts = trades['amount'].round(2).value_counts()
        most_common_amount = amount_counts.iloc[0] if len(amount_counts) > 0 else 0
        most_common_ratio = most_common_amount / len(trades)

        # 점수: 유사한 금액이 많거나, 특정 금액이 반복되면 높은 점수
        score = min((similar_ratio * 0.6 + most_common_ratio * 0.4) * 100, 100)

        evidence = {
            'similar_amount_pairs': len(similarities),
            'similar_ratio': round(similar_ratio, 3),
            'most_common_amount': round(trades['amount'].round(2).mode()[0], 2) if len(trades) > 0 else 0,
            'most_common_count': int(most_common_amount),
            'most_common_ratio': round(most_common_ratio, 3)
        }

        return score, evidence

    def _detect_symbol_concentration(self, trades: pd.DataFrame) -> Tuple[float, Dict]:
        """특정 심볼 집중 패턴 탐지"""
        symbol_dist = trades['symbol'].value_counts(normalize=True)

        # Herfindahl Index (집중도 지수)
        herfindahl = (symbol_dist ** 2).sum()

        top_symbol = symbol_dist.index[0]
        top_symbol_ratio = symbol_dist.iloc[0]

        # 점수: 집중도가 높을수록 높은 점수
        if top_symbol_ratio >= self.thresholds['symbol_concentration']:
            score = min(herfindahl * 100, 100)
        else:
            score = min(herfindahl * 70, 70)

        evidence = {
            'unique_symbols': trades['symbol'].nunique(),
            'top_symbol': top_symbol,
            'top_symbol_ratio': round(top_symbol_ratio, 3),
            'herfindahl_index': round(herfindahl, 3)
        }

        return score, evidence

    def _detect_high_frequency_pattern(self, trades: pd.DataFrame) -> Tuple[float, Dict]:
        """고빈도 단기 거래 패턴 탐지"""
        # 거래 간격 분석
        time_diffs = trades['time_diff_seconds'].dropna()

        if len(time_diffs) == 0:
            return 0, {'avg_interval': 0}

        avg_interval = time_diffs.mean()
        median_interval = time_diffs.median()
        min_interval = time_diffs.min()

        # 매우 짧은 간격의 거래 비율
        rapid_trades = (time_diffs < self.thresholds['min_trade_interval_seconds']).sum()
        rapid_ratio = rapid_trades / len(time_diffs)

        # 일일 거래 빈도
        trading_days = (trades['ts'].max() - trades['ts'].min()).days + 1
        daily_frequency = len(trades) / trading_days

        # 점수: 짧은 간격 거래가 많고, 일일 빈도가 높을수록 높은 점수
        interval_score = min((1 / (avg_interval + 1)) * 1000, 50)  # 간격 짧을수록 높음
        frequency_score = min(daily_frequency / 10 * 50, 50)  # 빈도 높을수록 높음

        score = interval_score + frequency_score

        evidence = {
            'avg_interval_seconds': round(avg_interval, 2),
            'median_interval_seconds': round(median_interval, 2),
            'min_interval_seconds': round(min_interval, 2),
            'rapid_trade_count': int(rapid_trades),
            'rapid_trade_ratio': round(rapid_ratio, 3),
            'daily_trade_frequency': round(daily_frequency, 2)
        }

        return score, evidence

    def _detect_no_price_impact(self, trades: pd.DataFrame) -> Tuple[float, Dict]:
        """가격 충격 없는 대량 거래 패턴"""
        no_impact_positions = 0
        total_positions = 0

        for position_id, pos_trades in trades.groupby('position_id'):
            pos_sorted = pos_trades.sort_values('ts')

            opens = pos_sorted[pos_sorted['openclose'] == 'OPEN']
            closes = pos_sorted[pos_sorted['openclose'] == 'CLOSE']

            if len(opens) > 0 and len(closes) > 0:
                total_positions += 1

                # OPEN 가격의 표준편차
                open_price_std = opens['price'].std()
                open_price_mean = opens['price'].mean()
                open_cv = (open_price_std / open_price_mean) if open_price_mean > 0 else 0

                # CLOSE 가격의 표준편차
                close_price_std = closes['price'].std()
                close_price_mean = closes['price'].mean()
                close_cv = (close_price_std / close_price_mean) if close_price_mean > 0 else 0

                # 가격 변동성이 매우 낮으면 (0.001% 미만) 가격 충격 없음
                if open_cv < 0.00001 and close_cv < 0.00001:
                    no_impact_positions += 1

        if total_positions == 0:
            return 0, {'no_impact_count': 0}

        no_impact_ratio = no_impact_positions / total_positions

        # 점수: 가격 충격 없는 거래가 많을수록 높은 점수
        score = min(no_impact_ratio * 100, 100)

        evidence = {
            'no_impact_position_count': no_impact_positions,
            'total_position_count': total_positions,
            'no_impact_ratio': round(no_impact_ratio, 3)
        }

        return score, evidence

    def analyze_account_detail(self, account_id: str) -> Dict:
        """특정 계정의 상세 분석"""
        account_trades = self.trade_df[self.trade_df['account_id'] == account_id].copy()

        if len(account_trades) == 0:
            return {'error': 'Account not found'}

        # 기본 정보
        analysis = {
            'account_id': account_id,
            'total_trades': len(account_trades),
            'trading_period': f"{account_trades['ts'].min()} ~ {account_trades['ts'].max()}",
            'symbols': account_trades['symbol'].unique().tolist(),
        }

        # 포지션별 상세 분석
        positions = []
        for position_id, pos_trades in account_trades.groupby('position_id'):
            pos_sorted = pos_trades.sort_values('ts')

            opens = pos_sorted[pos_sorted['openclose'] == 'OPEN']
            closes = pos_sorted[pos_sorted['openclose'] == 'CLOSE']

            position_info = {
                'position_id': position_id,
                'symbol': pos_sorted.iloc[0]['symbol'],
                'side': pos_sorted.iloc[0]['side'],
                'total_trades': len(pos_sorted),
                'open_count': len(opens),
                'close_count': len(closes)
            }

            if len(opens) > 0:
                position_info['first_open_time'] = opens['ts'].min()
                position_info['open_avg_price'] = (opens['price'] * opens['qty']).sum() / opens['qty'].sum()
                position_info['open_total_amount'] = opens['amount'].sum()

            if len(closes) > 0:
                position_info['last_close_time'] = closes['ts'].max()
                position_info['close_avg_price'] = (closes['price'] * closes['qty']).sum() / closes['qty'].sum()
                position_info['close_total_amount'] = closes['amount'].sum()

            if len(opens) > 0 and len(closes) > 0:
                duration = closes['ts'].min() - opens['ts'].max()
                position_info['duration_seconds'] = duration.total_seconds()
                position_info['duration_minutes'] = duration.total_seconds() / 60

                # 손익 계산
                side = opens.iloc[0]['side']
                open_avg = position_info['open_avg_price']
                close_avg = position_info['close_avg_price']

                if side == 'LONG':
                    profit_rate = ((close_avg - open_avg) / open_avg) * 100
                else:
                    profit_rate = ((open_avg - close_avg) / open_avg) * 100

                position_info['profit_rate'] = profit_rate
                position_info['abs_profit_rate'] = abs(profit_rate)

            positions.append(position_info)

        analysis['positions'] = pd.DataFrame(positions)

        # 거래 패턴 통계
        analysis['stats'] = {
            'avg_position_duration_minutes': analysis['positions']['duration_minutes'].mean() if 'duration_minutes' in analysis['positions'].columns else None,
            'median_position_duration_minutes': analysis['positions']['duration_minutes'].median() if 'duration_minutes' in analysis['positions'].columns else None,
            'avg_abs_profit_rate': analysis['positions']['abs_profit_rate'].mean() if 'abs_profit_rate' in analysis['positions'].columns else None,
            'near_zero_profit_count': (analysis['positions']['abs_profit_rate'] < 0.1).sum() if 'abs_profit_rate' in analysis['positions'].columns else 0,
        }

        return analysis

    def visualize_suspicious_account(self, account_id: str, analysis: Optional[Dict] = None):
        """의심 계정 시각화"""
        if analysis is None:
            analysis = self.analyze_account_detail(account_id)

        if 'error' in analysis:
            print(f"Error: {analysis['error']}")
            return

        account_trades = self.trade_df[self.trade_df['account_id'] == account_id].copy()
        positions_df = analysis['positions']

        fig = plt.figure(figsize=(18, 12))
        gs = fig.add_gridspec(3, 3, hspace=0.3, wspace=0.3)

        # 제목
        fig.suptitle(f'Wash Trading 의심 계정 상세 분석: {account_id}',
                     fontsize=16, fontweight='bold')

        # 1. 시간대별 거래 분포
        ax1 = fig.add_subplot(gs[0, 0])
        account_trades['hour'] = account_trades['ts'].dt.hour
        hourly_dist = account_trades.groupby('hour').size()
        ax1.bar(hourly_dist.index, hourly_dist.values, color='steelblue', alpha=0.7)
        ax1.set_xlabel('시간 (Hour)')
        ax1.set_ylabel('거래 횟수')
        ax1.set_title('시간대별 거래 분포')
        ax1.grid(True, alpha=0.3)

        # 2. 포지션 보유 시간 분포
        ax2 = fig.add_subplot(gs[0, 1])
        if 'duration_minutes' in positions_df.columns:
            ax2.hist(positions_df['duration_minutes'], bins=30,
                    color='coral', alpha=0.7, edgecolor='black')
            ax2.axvline(positions_df['duration_minutes'].median(),
                       color='red', linestyle='--', linewidth=2, label='중앙값')
            ax2.set_xlabel('포지션 보유 시간 (분)')
            ax2.set_ylabel('빈도')
            ax2.set_title('포지션 보유 시간 분포')
            ax2.legend()
            ax2.grid(True, alpha=0.3)

        # 3. 손익률 분포
        ax3 = fig.add_subplot(gs[0, 2])
        if 'profit_rate' in positions_df.columns:
            profit_rates = positions_df['profit_rate'].dropna()
            ax3.hist(profit_rates, bins=30, color='lightgreen',
                    alpha=0.7, edgecolor='black')
            ax3.axvline(0, color='red', linestyle='--', linewidth=2)
            ax3.set_xlabel('손익률 (%)')
            ax3.set_ylabel('빈도')
            ax3.set_title('포지션별 손익률 분포')
            ax3.grid(True, alpha=0.3)

        # 4. 거래 금액 분포
        ax4 = fig.add_subplot(gs[1, 0])
        ax4.hist(account_trades['amount'], bins=30,
                color='mediumpurple', alpha=0.7, edgecolor='black')
        ax4.set_xlabel('거래 금액')
        ax4.set_ylabel('빈도')
        ax4.set_title('거래 금액 분포')
        ax4.grid(True, alpha=0.3)

        # 5. 심볼별 거래 비율
        ax5 = fig.add_subplot(gs[1, 1])
        symbol_counts = account_trades['symbol'].value_counts()
        colors = plt.cm.Set3(range(len(symbol_counts)))
        ax5.pie(symbol_counts.values, labels=symbol_counts.index,
               autopct='%1.1f%%', colors=colors, startangle=90)
        ax5.set_title('심볼별 거래 비율')

        # 6. OPEN vs CLOSE 비율
        ax6 = fig.add_subplot(gs[1, 2])
        openclose_counts = account_trades['openclose'].value_counts()
        ax6.bar(openclose_counts.index, openclose_counts.values,
               color=['green', 'red'], alpha=0.7)
        ax6.set_xlabel('거래 유형')
        ax6.set_ylabel('횟수')
        ax6.set_title('OPEN vs CLOSE 비율')
        ax6.grid(True, alpha=0.3)

        # 7. 시계열 거래 패턴
        ax7 = fig.add_subplot(gs[2, :])
        for symbol in account_trades['symbol'].unique():
            symbol_trades = account_trades[account_trades['symbol'] == symbol]
            ax7.scatter(symbol_trades['ts'], symbol_trades['price'],
                       label=symbol, alpha=0.6, s=50)
        ax7.set_xlabel('시간')
        ax7.set_ylabel('거래 가격')
        ax7.set_title('시계열 거래 패턴')
        ax7.legend()
        ax7.grid(True, alpha=0.3)
        plt.xticks(rotation=45)

        plt.tight_layout()
        return fig

    def generate_report(self, result_df: pd.DataFrame, top_n: int = 10) -> str:
        """탐지 결과 리포트 생성"""
        if len(result_df) == 0:
            return "의심 계정이 발견되지 않았습니다."

        report = []
        report.append("=" * 80)
        report.append("🚨 WASH TRADING 탐지 리포트")
        report.append("=" * 80)
        report.append(f"\n총 의심 계정 수: {len(result_df)}")
        report.append(f"분석 대상 전체 계정 수: {self.trade_df['account_id'].nunique()}")
        report.append(f"의심 계정 비율: {len(result_df) / self.trade_df['account_id'].nunique() * 100:.2f}%")

        report.append("\n" + "-" * 80)
        report.append(f"상위 {min(top_n, len(result_df))}개 고위험 계정:")
        report.append("-" * 80)

        for idx, row in result_df.head(top_n).iterrows():
            report.append(f"\n[{idx+1}] 계정: {row['account_id']}")
            report.append(f"    의심 점수: {row['suspicious_score']:.1f}/100")
            report.append(f"    총 거래 횟수: {row['total_trades']}")
            report.append(f"    거래 심볼 수: {row['unique_symbols']}")
            report.append(f"    거래 기간: {row['trading_period_days']}일")
            report.append(f"    세부 점수:")
            report.append(f"      - 빠른 왕복거래: {row['rapid_round_trips']:.1f}")
            report.append(f"      - 제로 수익: {row['zero_profit']:.1f}")
            report.append(f"      - 동일 금액 반복: {row['identical_amount']:.1f}")
            report.append(f"      - 심볼 집중: {row['symbol_concentration']:.1f}")
            report.append(f"      - 고빈도 거래: {row['high_frequency']:.1f}")
            report.append(f"      - 가격 충격 없음: {row['no_price_impact']:.1f}")

        report.append("\n" + "=" * 80)

        return "\n".join(report)

    def export_results(self, result_df: pd.DataFrame,
                      output_path: str = 'wash_trading_suspects.csv'):
        """결과를 CSV로 저장"""
        if len(result_df) == 0:
            print("저장할 의심 계정이 없습니다.")
            return

        # evidences 컬럼 제외하고 저장
        export_df = result_df.drop('evidences', axis=1)
        export_df.to_csv(output_path, index=False, encoding='utf-8-sig')
        print(f"✅ 결과가 {output_path}에 저장되었습니다.")


# ============================================================================
# 실행 예시
# ============================================================================

if __name__ == "__main__":
    # 1. Detector 초기화
    detector = WashTradingDetector(data)

    # 2. Wash Trading 탐지 실행
    suspicious_accounts = detector.detect_all_patterns()

    # 3. 결과가 있는 경우
    if len(suspicious_accounts) > 0:
        # 리포트 생성 및 출력
        report = detector.generate_report(suspicious_accounts, top_n=10)
        print("\n" + report)

        # 결과 저장
        detector.export_results(suspicious_accounts)

        # 상위 3개 계정 상세 분석 및 시각화
        print("\n" + "=" * 80)
        print("상위 의심 계정 상세 분석")
        print("=" * 80)

        for idx, account_id in enumerate(suspicious_accounts.head(3)['account_id']):
            print(f"\n[{idx+1}] {account_id} 분석 중...")

            # 상세 분석
            analysis = detector.analyze_account_detail(account_id)

            # 통계 출력
            print(f"총 거래: {analysis['total_trades']}")
            print(f"거래 기간: {analysis['trading_period']}")
            print(f"거래 심볼: {', '.join(analysis['symbols'])}")
            print(f"평균 포지션 보유 시간: {analysis['stats']['avg_position_duration_minutes']:.2f}분")
            print(f"평균 절대 수익률: {analysis['stats']['avg_abs_profit_rate']:.3f}%")
            print(f"제로 수익 포지션 수: {analysis['stats']['near_zero_profit_count']}")

            # 시각화
            fig = detector.visualize_suspicious_account(account_id, analysis)
            plt.savefig(f'wash_trading_analysis_{account_id}.png', dpi=150, bbox_inches='tight')
            print(f"✅ 시각화 저장: wash_trading_analysis_{account_id}.png")

        plt.show()

    else:
        print("\n✅ Wash Trading 의심 계정이 발견되지 않았습니다.")
```
